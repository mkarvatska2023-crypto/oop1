# Перша програма на ООП

```python

class MyName:
    """Опис класу / Документація
    """
    total_names = 0 #Class Variable

    def __init__(self, name=None) -> None:
        """Ініціалізація класу
        """
        self.name = name if name is not None else self.anonymous_user().name #Class attributes / Instance variables
        MyName.total_names += 1 #modify class variable
        self.my_id = self.total_names

    @property
    def whoami(self) -> str: 
        """Class property
        return: повертаємо імя 
        """
        return f"My name is {self.name}"
    
    @property
    def my_email(self) -> str:
        """Class property
        return: повертаємо емейл
        """
        return self.create_email()
    
    def create_email(self) -> str:
        """Instance method
        """
        return f"{self.name}@itcollege.lviv.ua"

    @classmethod
    def anonymous_user(cls):
        """Classs method
        """
        return cls("Anonymous")
    
    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        """Static method
        """
        return f"You say: {message}"


print("Розпочинаємо створювати обєкти!")

names = ("Bohdan", "Marta", None)
all_names = {name: MyName(name) for name in names}

for name, me in all_names.items():
    print(f"""{">*<"*20}
This is object: {me} 
This is object attribute: {me.name} / {me.my_id}
This is {type(MyName.whoami)}: {me.whoami} / {me.my_email}
This is {type(me.create_email)} call: {me.create_email()}
This is static {type(MyName.say_hello)} with defaults: {me.say_hello()} 
This is class variable {type(MyName.total_names)}: from class {MyName.total_names} / from object {me.total_names}
{"<*>"*20}""")

print(f"We are done. We create {me.total_names} names! ??? Why {MyName.total_names}?")
```

- мій приклад

```python
class MyName:
    """Клас для збереження імені користувача та створення email."""
    
    total_names = 0  # змінна класу

    def __init__(self, name=None) -> None:
        """Ініціалізація об'єкта."""
        # Якщо ім’я не передано — створюємо анонімного користувача
        self.name = name if name else self.anonymous_user().name

        MyName.total_names += 1
        self.my_id = MyName.total_names

    @property
    def whoami(self) -> str:
        """Повертає рядок з іменем."""
        return f"My name is {self.name}"

    @property
    def my_email(self) -> str:
        """Повертає email користувача."""
        return self.create_email()

    def create_email(self) -> str:
        """Створює email."""
        return f"{self.name}@itcollege.lviv.ua"

    @classmethod
    def anonymous_user(cls):
        """Створює анонімного користувача."""
        return cls("Anonymous")

    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        """Повертає вітальне повідомлення."""
        return f"You say: {message}"


# === Створення об'єктів ===

print("Починаємо створювати об’єкти!")

names = ("Maria", "Anna", None)
all_names = {name: MyName(name) for name in names}

for name, obj in all_names.items():
    print(f"""{">*<"*20}
Object: {obj}
Name: {obj.name} | ID: {obj.my_id}
Properties: {obj.whoami} | {obj.my_email}
Email method: {obj.create_email()}
Static method: {obj.say_hello()}
Class variable: {MyName.total_names} (class) / {obj.total_names} (object)
{"<*>"*20}""")

print(f"Усього створено {MyName.total_names} об’єктів!")
```

```python
from datetime import datetime
class MyAnimals:
    max_age = 20
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"


a1 = MyAnimals("Кіт", "Мурка", 5)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
a1.add_one_year()
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
```

- мій приклад

```python
from datetime import datetime

class MyAnimals:
    max_age = 20  # максимальний вік для будь-якої тваринки

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_untill(self):
        """Розрахунок року, до якого житиме тваринка"""
        return datetime.today().year - self.age + self.max_age

    def add_one_year(self):
        """Додаємо тваринці ще один прожитий рік"""
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік!"


a1 = MyAnimals("Кіт", "Мурка", 5)

print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року.")
a1.add_one_year()
print(f"Після ще одного року мій {a1.animals} буде зі мною до {a1.live_untill} року.")
```

```python

a = 2
b = 5
c = "C"

print(a+b, a - int(1),  a * b, str(a) + c)

print(a.__add__(b), a.__sub__(int(1)), a.__mul__(b))

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"Мій {self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            self.age += obj
            return f"Мій {self.name} постарів на {obj} років, тепер йому {self.age}"
        return NotImplementedError
    
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"Старшим є {self.name}"
            return f"{self.name} не є старшим за {obj.name}"
        return NotImplementedError

    def __len__(self) -> int:
        print("Ми зайшли в метон len, та знаходимо довжину Імені тваринки")
        return len(self.name)
            
        

a = MyAnimals("Собака", "Рекс", 2)
b = MyAnimals("Кіт", "Мурзик", 4)
print(a - b)
print(a - 1)
print(a - float(1))

print(a > b)

print(b > a)

len(a)
```

- приклад з лекції

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.description = f"|Назва: {self.title};\n|Автор: {self.author};"
    
    def __str__(self):
        return self.description
```

### Клас бібліотеки
- приклад з лекції

```python
class Library:
    def __init__(self):
        self.books: list[Book] = []
    
    def add_book(self, book: Book):
        print(f"Додаємо книгу: {book.title}")
        self.books.append(book)
    
    @property
    def list_books(self):
        return self.books
```

- приклад з лекції

```python
it_collage_library = Library()
for name in ["Хірург", "Майстер і Маргарита", "1984", "Гаррі Поттер", "Історія Львова"]:
    book = Book(name, "Unknown")
    it_collage_library.add_book(book)

print(">>> Список книг у бібліотеці:")
for book in it_collage_library.list_books:
    print(f"{book}\n|Статус: Доступна\n---")
```

### клас студента
- приклад з лекції

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.student_id = id(self)
        self.description = f"|Ім'я: {self.name};\n|ID студента: {self.student_id};"
        self.borrowed_books: list[Book] = []
    
    # Додаємо метод перегляду які книжки зараз має студент
    @property
    def list_borrowed_books(self):
        for b in self.borrowed_books:
            print(f"Взято книгу: {b}")
    
    def __str__(self):
        return self.description
```

```python
class Library:
    def __init__(self):
        self.books: list[Book] = []
        self.students: list[Student] = []
    
    def add_book(self, book: Book):
        """Метод для додавання книги до бібліотеки"""
        print(f"Додаємо книгу: {book.title}")
        self.books.append(book)
    
    @property
    def list_books(self):
        return self.books
    
    @property
    def list_students(self):
        return self.students
    
    def register_student(self, student: Student):
        """Метод для реєстрації студента в бібліотеці"""
        print(f"Реєструємо студента: {student.name}")
        self.students.append(student)
    
    # Додаємо метод для видачі книги студенту
    def lend_book(self, book_title: str, student_id: int):
        # Знаходимо книгу за назвою
        book_to_lend = next((book for book in self.books if book.title == book_title), None)
        if not book_to_lend:
            print(f"Книга '{book_title}' не знайдена в бібліотеці.")
            return
        
        # Знаходимо студента за ID
        student = next((s for s in self.students if s.student_id == student_id), None)
        if not student:
            print(f"Студент з ID '{student_id}' не зареєстрований в бібліотеці.")
            return
        
        # Видаємо книгу студенту
        student.borrowed_books.append(book_to_lend)
        self.books.remove(book_to_lend)
        print(f"Книга '{book_title}' видана студенту '{student.name}'.")
```

- приклад з лекції

```python
it_collage_library = Library()
for name in ["Хірург", "Майстер і Маргарита", "1984", "Гаррі Поттер", "Історія Львова"]:
    book = Book(name, "Unknown")
    it_collage_library.add_book(book)

for student in ["Богдан", "Анна", "Марія", "Руслан"]:
    new_student = Student(student)
    it_collage_library.register_student(new_student)

print(">>> Список книг у бібліотеці:")
for book in it_collage_library.list_books:
    print(f"{book}\n|Статус: Доступна\n---")

print(">>> Список зареєстрованих студентів:")
for student in it_collage_library.list_students:
    print(f"{student}\n---")
```
